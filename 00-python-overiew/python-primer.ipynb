# 4. Functions

##### Problem: program to detect if a number is prime or not

```python
n = int(input("Enter a number: "))
prime = True
for i in range(2, n):
    if(n % i == 0):
        prime = False
        break
if prime == True:
    print("It is a prime number")
else:
    print("It is not a prime number")
```

```python
def checkprime(n: int) -> bool:
    for i in range(2, n):
        if(n % i == 0):
            return False
    return True
```

```python
primes = []
for n in range(1, 101):
    if(checkprime(n)):
        primes.append(n)
```

```python
primes
```

```python
[n for n in range(1, 101) if checkprime(n)]
```
