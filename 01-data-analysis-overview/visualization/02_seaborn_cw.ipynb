### Import Seaborn

```python
import seaborn as sns
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv(r"tips.csv")
df.head()
```

### Categorical Plots

### Bar Plot

```python
sns.barplot(x=df["gender"], y=df["total_bill"], data=df)
```

```python
sns.barplot(x=df["gender"], y=df["total_bill"], data=df, estimator=np.std)
```

### Count Plot

```python
sns.countplot(x=df["gender"], data=df)
```

### Box Plots

```python
df.head()
```

```python
sns.boxplot(x=df['gender'],y=df['total_bill'],data=df)
```

```python
sns.boxplot(x=df['day'],y=df['total_bill'],data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df, hue="gender", palette="coolwarm")
```

### Distribution Plots

```python
sns.histplot(df["total_bill"])
```

```python
sns.histplot(df["total_bill"], kde=False)
```

### Joint Plots

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

### Pair Plots

```python
sns.pairplot(df, hue="gender", palette='coolwarm')
```

### Heat Maps

```python
df.head()
```

```python
df[['total_bill', 'tip']].corr()
```

```python
sns.heatmap(df[['total_bill', 'tip']].corr())
```

```python
sns.heatmap(df[['total_bill', 'tip']].corr(), cmap='coolwarm', annot=True)
```
